```python
df = df.dropna()
```

```python
df=df.fillna(0)
```

```python
df=df.drop_duplicates()
```

```python
df["Date"] = pd.to_datetime(df["Date"], errors="coerce")
```

```python
df["TotalPrice"] = pd.to_numeric(df["TotalPrice"], errors="coerce")
```

```python
df.info()
df.isnull().sum()
```

```python
df.to_excel("cleaned_data.xlsx", index=False)
```
